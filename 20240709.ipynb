```python
import torch
from transformers import AutoModelForSpeechSeq2Seq, AutoProcessor, pipeline
from datasets import load_dataset


device = "cuda:0" if torch.cuda.is_available() else "cpu"
torch_dtype = torch.float16 if torch.cuda.is_available() else torch.float32

model_id = "openai/whisper-large-v3"

model = AutoModelForSpeechSeq2Seq.from_pretrained(
    model_id, torch_dtype=torch_dtype, low_cpu_mem_usage=True, use_safetensors=True
)
model.to(device)

processor = AutoProcessor.from_pretrained(model_id)

pipe = pipeline(
    "automatic-speech-recognition",
    model=model,
    tokenizer=processor.tokenizer,
    feature_extractor=processor.feature_extractor,
    max_new_tokens=128,
    chunk_length_s=30,
    batch_size=16,
    return_timestamps=True,
    torch_dtype=torch_dtype,
    device=device,
)

dataset = load_dataset("distil-whisper/librispeech_long", "clean", split="validation")
sample = dataset[0]["audio"]

result = pipe(sample)
print(result["text"])
```

```python
import gradio as gr
import librosa
import torch
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer

# 加载预训练的Wav2Vec2模型和分词器
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")

def transcribe_audio(audio):
    # 使用 librosa 加载音频文件，指定采样率为 16000 Hz
    audio_data, sampling_rate = librosa.load(audio, sr=16000)

    # 将音频数据转换为输入特征
    input_values = tokenizer(audio_data, return_tensors="pt", padding="longest").input_values

    # 使用模型进行语音识别
    with torch.no_grad():
        logits = model(input_values).logits

    # 解码输出为文本
    predicted_ids = torch.argmax(logits, dim=-1)
    transcription = tokenizer.batch_decode(predicted_ids)[0]
    return transcription

with gr.Blocks() as demo:
    gr.Markdown("please use English to text. if you utilize Chinese to translate, the outcome will be extremely strange.")

    with gr.Tab("音檔"):
        with gr.Column():
            audio_input = gr.Audio(type="filepath", label="請上傳音檔")
            text_output = gr.Textbox(label="轉錄結果")
            transcribe_button = gr.Button("轉換")

    transcribe_button.click(fn=transcribe_audio, inputs=audio_input, outputs=text_output)

# 启动Gradio应用
if __name__ == "__main__":
    demo.launch()
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# 加载 GPT2 分词器和模型
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

# 使用前面步骤得到的转录文本
transcriptioned = 'Baverage store'

# 分词和编码文本
input_ids = tokenizer.encode(transcriptioned, return_tensors="pt")

# 生成文本
output = model.generate(input_ids, max_length=60, num_return_sequences=1, no_repeat_ngram_size=2, early_stopping=True)

# 解码生成的文本
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

print('Question:', transcriptioned)
print("Generated text:", generated_text)
```

```python
import gradio as gr
import librosa
import torch
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
import os
import google.generativeai as genai

# 设置 API 密钥
api_key = 'AIzaSyCw-nt7M1gSQ_m8xFlTRyimr68ZVIhBxdU'
os.environ['API_KEY'] = api_key

# 配置 Gemini API
genai.configure(api_key=os.environ['API_KEY'])

# 选择适合的模型
gemini_model = genai.GenerativeModel('gemini-1.5-flash')

# 加载预训练的Wav2Vec2模型和分词器
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")

def transcribe_audio(audio):
    # 使用 librosa 加载音频文件，指定采样率为 16000 Hz
    audio_data, sampling_rate = librosa.load(audio, sr=16000)

    # 将音频数据转换为输入特征
    input_values = tokenizer(audio_data, return_tensors="pt", padding="longest").input_values

    # 使用模型进行语音识别
    with torch.no_grad():
        logits = model(input_values).logits

    # 解码输出为文本
    predicted_ids = torch.argmax(logits, dim=-1)
    transcription = tokenizer.batch_decode(predicted_ids)[0]
    return transcription

def generate_response(transcription):
    # 使用 Gemini API 生成内容
    response = gemini_model.generate_content(transcription)
    return response.text

def stop_app():
    global stop_generation
    stop_generation = True

with gr.Blocks() as demo:
    gr.Markdown("Response to your question")

    with gr.Tab("音檔"):
        with gr.Column():
            audio_input = gr.Audio(type="filepath", label="請上傳音檔")
            text_output = gr.Textbox(label="轉錄結果")
            transcribe_button = gr.Button("轉換")

            gemini_output = gr.Textbox(label="Gemini 回應")
            gemini_button = gr.Button("生成回應")

            stop_button = gr.Button("強制結束")

    transcribe_button.click(fn=transcribe_audio, inputs=audio_input, outputs=text_output)
    gemini_button.click(fn=generate_response, inputs=text_output, outputs=gemini_output)
    stop_button.click(fn=stop_app)

# 启动Gradio应用
if __name__ == "__main__":
    demo.launch()

if __name__ == "__main__":
    demo.launch(server_name="0.0.0.0", server_port=7860, timeout=300)
```
